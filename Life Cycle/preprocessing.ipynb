3.1/ Preprocessing on Training data
</h3>


```python
# Define Numerical features in training data
num_cols = X_train.select_dtypes(include=['int64','float64']).columns
print(num_cols)
```

```python
# Define Categorical features in training data
cat_cols = X_train.select_dtypes(include=['object','bool']).columns
print(cat_cols)
```

```python
# Fill Null With mean in numerical features
for col in num_cols:
    X_train[col] = X_train[col].fillna(X_train[col].mean())
```

```python
# Fill Null With mode in categorical features
for col in cat_cols:
    X_train[col] = X_train[col].fillna(X_train[col].mode()[0])
```

```python
# Define Categorical features in testing data
cat_cols_test = X_test.select_dtypes(include=['object','bool']).columns
print(cat_cols_test)
```

```python
# Fill Null With mean in numerical features
for col in num_cols_test:
    X_test[col] = X_test[col].fillna(X_test[col].mean())
```

```python
# Fill Null With mean in categorical features
for col in cat_cols_test:
    X_test[col] = X_test[col].fillna(X_test[col].mode()[0])
```

Step 1: Apply One-Hot Encoding on training data
</h5>

```python
# Convert categorical columns in X_train into numeric using One-Hot Encoding
X_train = pd.get_dummies(X_train, drop_first=True)
```

<h5 style="
background-color:#1e1e1e;
color:#FFD700;
padding:6px 10px;

">
Step 2: Apply One-Hot Encoding on testing data
</h5>

```python
# Convert categorical columns in X_test into numeric using the same encoding method
X_test = pd.get_dummies(X_test, drop_first=True)
```


<h5 style="
background-color:#1e1e1e;
color:#FFD700;
padding:6px 10px;
border-radius:6px;
">
Step 3: Align columns between train and test
</h5>

```python
# Align X_train and X_test to have the same columns
# Missing columns in X_test will be filled with 0
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```
